```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('PySpark').getOrCreate()
```

```python
df = spark.read.option('header','true').csv('articles.csv')
```

```python
%time

from pyspark.sql.functions import translate,upper,lower
df.withColumn('prod_name', lower(translate(upper('prod_name'), 'S', 'R'))).select('prod_name').show()
df.withColumn('product_type_name', lower(translate(upper('product_type_name'), 'S', 'R'))).select('product_type_name').show()
df.withColumn('product_group_name', lower(translate(upper('product_group_name'), 'S', 'R'))).select('product_group_name').show()

#df.show()
```
